```python
# packages

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
# loading data

iris = datasets.load_iris()

data = pd.DataFrame(iris.data, columns=iris.feature_names)

X = np.array(data)

data
```

```python
kmeans = KMeans(n_clusters=3)

kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_
print(y_kmeans)
print(centers)
```

```python
predict_true = pd.DataFrame(iris.target, columns=['real'])
predict_true['predict'] = pd.Series(y_kmeans)

errors = len(predict_true.loc[predict_true['real'] != predict_true['predict']])
good = len(predict_true.loc[predict_true['real'] == predict_true['predict']])
print(errors)
print(good)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200);
```
